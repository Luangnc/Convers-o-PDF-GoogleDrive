**Código para importar arquivo PDF para o Google Drive**

```python
import requests

url = "Yourlink"
r = requests.get(url)

with open("/content/arquivo.pdf", "wb") as f:
    f.write(r.content)
```

```python
from google.colab import drive
drive.mount('/content/drive')
```
